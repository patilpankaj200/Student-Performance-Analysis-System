```python
df["average_score"] = (
    df["math score"] +
    df["reading score"] +
    df["writing score"]
) / 3
```

```python
df.columns
```

```python
df[["math score", "reading score", "writing score", "average_score"]].head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.barplot(x="gender", y="average_score", data=df)
plt.title("Average Score by Gender")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(df[["math score","reading score","writing score","average_score"]].corr(),
            annot=True,
            cmap="coolwarm")
plt.title("Correlation Between Subjects")
plt.show()
```

```python
sns.barplot(x="lunch", y="average_score", data=df)
plt.title("Average Score by Lunch Type")
plt.show()
```
